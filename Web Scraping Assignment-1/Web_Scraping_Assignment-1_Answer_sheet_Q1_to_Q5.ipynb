# Answer 1 - Header tags from wikipedia.org

Importing Required Libraries

```python
from bs4 import BeautifulSoup
import requests
```

Sending get request to the webpage server to get the source code of the page.

```python
page1 = requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page1
```

I got the response 200 means "Success", I can scrap the data from here.

Page content

```python
soup1 = BeautifulSoup(page1.content)
soup1
```

Scrapping First Header

```python
header1 = soup1.find('h2', class_="mp-h2")
header1
```

```python
header1.text
```

Scraping Multiple Titles

```python
header = []

for i in soup1.find_all('h2', class_="mp-h2"):
    header.append(i.text)

header
```

```python
# printing Length
print(len(header))
```

```python
# Making dataframe

import pandas as pd
df1 = pd.DataFrame({'header':header})
df1
```

# Answer 2 - IMDB’s Top rated 50 movies’

Sending request

```python
page2 = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
page2
```

Page content

```python
soup2 = BeautifulSoup(page2.content)
soup2
```

Scrapping First movie name

```python
movie = soup2.find('td', class_="titleColumn")
movie
```

```python
movie.text
```

Spliting text

```python
movie.text.split('\n')[2]
```

```python
rating = soup2.find('td', class_="ratingColumn imdbRating")
rating
```

```python
rating.text
```

```python
rating.text.split('\n')[1]
```

Scrapping First movie year of release (YOR)

```python
YOR = soup2.find('span', class_="secondaryInfo")
YOR
```

```python
YOR.text
```

Scrapping Multiple movie name

```python
MovieName = []

for i in soup2.find_all('td', class_="titleColumn"):
    MovieName.append(i.text.split('\n')[2])

MovieName
```

Scrapping Multiple movie rating

```python
MovieRating = []

for i in soup2.find_all('td', class_="ratingColumn imdbRating"):
    MovieRating.append(i.text.split('\n')[1])

MovieRating
```

Scrapping Multiple movie year of release

```python
MovieYOR = []

for i in soup2.find_all('span', class_="secondaryInfo"):
    MovieYOR.append(i.text)

MovieYOR
```

```python
# printing Length
print(len(MovieName),len(MovieRating),len(MovieYOR))
```

```python
# Making dataframe

movie_df = pd.DataFrame({'Name':MovieName, 'rating':MovieRating, 'year of release':MovieYOR})
movie_df
```
