```python
x = [5,8,10,2,6,12]
y = [12,16,6,10,14,6]
plt.bar(x,y)
plt.title('Bar graph')
plt.ylable('Y axis')
plt.xlable('X axis')
plt.show()
```

```python
# colour = 'g','b','y','r',   align ='center', 'edge'

plt.bar(x,y, color = 'g', align = 'edge')
plt.title('Bar graph')
plt.ylable('Y axis')
plt.xlable('X axis')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x=['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu']
y=[5,6.7,4,6,2,4.9,1.8]
```

```python
ax=sns.stripplot(x,y)
ax.set(xlabel='Days' ,ylabel='Amoun_spend')
plt.title('My first graph')
plt.show()
```

```python
# Load the data
ds=pd.read_csv('emptest.csv')
print(ds)
print(ds.head())
```

```python
# Create violinplot
sns.violinplot(x="Salary", data=ds)

# show the plot
plt.show()
```

```python
sns.violinplot(x = 'Age', data= ds)
```

```python
dsiris = pd.read_csv('iris.csv')
print(dsiris)
print(dsiris.head())
```

```python
sns.set(style = "whitegrid")
ax=sns.stripplot(x='class' , y='sepal length', data=dsiris)
plt.title('Graph')
plt.show
```

```python
ax=sns.swarmplot(x='class', y='sepal length', data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='petal length', y='class', data=dsiris)
plt.title('Graph')
plt.show()
```
